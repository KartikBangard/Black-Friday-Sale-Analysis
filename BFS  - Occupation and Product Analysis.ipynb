```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
sns.countplot(x = df['Stay_In_Current_City_Years'])
```

```python
df.groupby('Stay_In_Current_City_Years').size().plot(kind = 'pie',autopct = '%.1f')
```

```python
sns.countplot(x = df['Stay_In_Current_City_Years'], hue = "Gender", data = df)
```

```python
sns.countplot(x = df['Stay_In_Current_City_Years'], hue = "Marital_Status", data = df)
```

```python
sns.countplot(x = df['Stay_In_Current_City_Years'], hue = "City_Category", data = df)
```

```python
sns.countplot(x = df['City_Category'], hue = "Age", data = df)
```

```python
df.groupby('Stay_In_Current_City_Years').sum()['Purchase'].plot(kind = 'bar',figsize = (8,4))
```

```python
df.groupby('Occupation').size().sort_values().plot(kind = 'bar',figsize = (8,4))
```

```python
df.groupby('Occupation').sum()['Purchase'].sort_values().plot(kind = 'bar',figsize = (8,4))
```

```python
df.groupby('Occupation')['Purchase'].mean().sort_values().plot(kind = 'bar',figsize = (8,4))
```

```python
sns.countplot(x = 'Occupation',hue = 'Marital_Status',data = df)
```

```python
sns.countplot(x = 'Occupation',hue = 'Gender',data = df)
```

```python
df.groupby('Occupation').nunique()['Product_ID'].sort_values().plot(kind = 'bar',figsize = (8,4))
```

```python
df.groupby('Product_Category_1').size().sort_values().plot(kind = 'bar')
```

```python
df.groupby('Product_Category_1')['Purchase'].sum().sort_values().plot(kind = 'bar')
```

```python
df.groupby('Product_Category_1')['Purchase'].mean().sort_values().plot(kind = 'bar')
```

```python
df.groupby('Product_ID')['Purchase'].mean().nlargest(10).sort_values().plot(kind = 'bar',figsize = (6,3))
```

```python
df.groupby('Product_ID')['Purchase'].sum().nlargest(10).sort_values().plot(kind = 'bar',figsize = (6,3))
```

```python
df.groupby('Product_ID').size().nlargest(10).sort_values().plot(kind = 'bar',figsize = (6,3))
```

```python
sns.countplot(x = 'Product_Category_1',hue = 'Marital_Status',data = df)
```
