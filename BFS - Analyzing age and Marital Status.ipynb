```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
df.groupby('Age').size().plot(kind = 'bar',figsize = (8,6),title = "Distribution by Age",color = 'coral')
```

```python
l = []
for i in df['Age'].unique():
    l.append([i,df[df['Age'] == i]['Product_ID'].nunique()])
data = pd.DataFrame(l,columns = ['Age',"Products"])
```

```python
data.plot.bar(x = 'Age',figsize = (6,4))
```
